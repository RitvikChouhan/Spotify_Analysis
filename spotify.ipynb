```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('spotify_tracks.csv')
df_tracks.head()
```

```python
# null values

df_tracks.isnull().sum()
```

```python
df_tracks.info()
```

```python
# 10 least popular songs

sorted_df = df_tracks.sort_values('popularity', ascending = True).head(10)
sorted_df
```

```python
df_tracks.describe().transpose()
```

```python
# Top 10 popular songs

most_popular = df_tracks.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
df_tracks[["artists"]].iloc[18]
```

```python
# converting the duration_ms to seconds

df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace = True, axis = 1)
```

```python
df_tracks
```

```python
df_tracks = df_tracks.reset_index(drop=True)
df_tracks
```

```python
df_tracks.duration.head(10)
```

```python
print(df_tracks.info())
```

```python
corr_df = df_tracks.select_dtypes(include=[np.number]).corr()
```

```python
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot=True, cmap="inferno", vmin=-1, vmax=1, linecolor="Black", linewidth=1)
plt.show()
```

```python
# Filter only numeric columns
numeric_df_tracks = df_tracks.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numeric_df_tracks.corr()

# Display the correlation matrix
print(correlation_matrix)
```
